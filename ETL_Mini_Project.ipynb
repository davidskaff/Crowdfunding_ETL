### Create the Category and Subcategory DataFrames
---
**Create a Category DataFrame that has the following columns:**
- A "category_id" column that is numbered sequential form 1 to the length of the number of unique categories.
- A "category" column that has only the categories.

Export the DataFrame as a `category.csv` CSV file.

**Create a SubCategory DataFrame that has the following columns:**
- A "subcategory_id" column that is numbered sequential form 1 to the length of the number of unique subcategories.
- A "subcategory" column that has only the subcategories. 

Export the DataFrame as a `subcategory.csv` CSV file.

```python
# Get the crowdfunding_info_df columns.
crowdfunding_info_df_columns = crowdfunding_info_df.columns
print(crowdfunding_info_df_columns)
```

```python
# Split the 'category & sub-category' column into two separate columns.
crowdfunding_info_df[['category', 'subcategory']] = crowdfunding_info_df['category & sub-category'].str.split('/', expand=True)

# Remove any leading or trailing whitespace from the new columns.
crowdfunding_info_df['category'] = crowdfunding_info_df['category'].str.strip()
crowdfunding_info_df['subcategory'] = crowdfunding_info_df['subcategory'].str.strip()

# Display the DataFrame to verify the changes.
print(crowdfunding_info_df.head().to_string())
```

```python
# Get the unique categories and subcategories in separate lists.
categories = crowdfunding_info_df['category'].unique().tolist()
subcategories = crowdfunding_info_df['subcategory'].unique().tolist()

# Print the lists.
print("Categories:", categories)
print("Subcategories:", subcategories)
```

```python
# Get the number of distinct values in the categories and subcategories lists.
print(len(categories))
print(len(subcategories))
```

```python
# Create numpy arrays from 1-9 for the categories and 1-24 for the subcategories.
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
# Use a list comprehension to add "cat" to each category_id.
cat_ids = ["cat" + str(i) for i in range(1, len(categories) + 1)]

# Use a list comprehension to add "subcat" to each subcategory_id.
scat_ids = ["subcat" + str(i) for i in range(1, len(subcategories) + 1)]

# Print the lists.
print("Category IDs:", cat_ids)
print("Subcategory IDs:", scat_ids)
```

```python
# Create a category DataFrame with the category_id array as the category_id and categories list as the category name.
category_df = pd.DataFrame({
    'category_id': cat_ids,
    'category': categories
})

# Create a subcategory DataFrame with the subcategory_id array as the subcategory_id and subcategories list as the subcategory name.
subcategory_df = pd.DataFrame({
    'subcategory_id': scat_ids,
    'subcategory': subcategories
})

# Display the DataFrames to verify the changes.
print(category_df.head())
print(subcategory_df.head())
```

```python
# Export categories_df and subcategories_df as CSV files.
category_df.to_csv("Resources/category.csv", index=False)
subcategory_df.to_csv("Resources/subcategory.csv", index=False)
```

### Campaign DataFrame
----
**Create a Campaign DataFrame that has the following columns:**
- The "cf_id" column.
- The "contact_id" column.
- The “company_name” column.
- The "blurb" column is renamed as "description."
- The "goal" column.
- The "goal" column is converted to a `float` datatype.
- The "pledged" column is converted to a `float` datatype. 
- The "backers_count" column. 
- The "country" column.
- The "currency" column.
- The "launched_at" column is renamed as "launch_date" and converted to a datetime format. 
- The "deadline" column is renamed as "end_date" and converted to a datetime format.
- The "category_id" with the unique number matching the “category_id” from the category DataFrame. 
- The "subcategory_id" with the unique number matching the “subcategory_id” from the subcategory DataFrame.
- And, create a column that contains the unique four-digit contact ID number from the `contact.xlsx` file.
 

Then export the DataFrame as a `campaign.csv` CSV file.


```python
# Create a copy of the crowdfunding_info_df DataFrame name campaign_df. 
campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
# Rename the blurb, launched_at, and deadline columns.
crowdfunding_info_df = crowdfunding_info_df.rename(columns={
    'blurb': 'description',
    'launched_at': 'launch_date',
    'deadline': 'end_date'
})

# Display the DataFrame to verify the changes.
print(crowdfunding_info_df.head().to_string())
```

```python
# Convert the goal and pledged columns to a `float` data type.
crowdfunding_info_df['goal'] = crowdfunding_info_df['goal'].astype(float)
crowdfunding_info_df['pledged'] = crowdfunding_info_df['pledged'].astype(float)

# Display the DataFrame to verify the changes.
print(crowdfunding_info_df.head().to_string())
```

```python
# Check the datatypes
print(crowdfunding_info_df.dtypes)
```

```python
# Format the launched_date and end_date columns to datetime format
from datetime import datetime as dt
crowdfunding_info_df['launch_date'] = pd.to_datetime(crowdfunding_info_df['launch_date'], unit='s')
crowdfunding_info_df['end_date'] = pd.to_datetime(crowdfunding_info_df['end_date'], unit='s')

# Display the DataFrame to verify the changes.
print(crowdfunding_info_df.head().to_string())
```

```python
# Merge the campaign_df with the category_df on the "category" column.
campaign_merged_df = pd.merge(crowdfunding_info_df, category_df, how='left', on='category')

# Merge the resulting DataFrame with the subcategory_df on the "subcategory" column.
campaign_merged_df = pd.merge(campaign_merged_df, subcategory_df, how='left', on='subcategory')

# Display the last 10 rows of the merged DataFrame.
print(campaign_merged_df.tail(10).to_string())
```

```python
# Drop unwanted columns
# List of columns to drop.
columns_to_drop = ['staff_pick', 'spotlight', 'category & sub-category', 'category', 'subcategory']

# Drop unwanted columns.
campaign_cleaned = campaign_merged_df.drop(columns=columns_to_drop)

# Remove the time part from the 'launch_date' and 'end_date' columns.
campaign_cleaned['launch_date'] = campaign_cleaned['launch_date'].dt.date
campaign_cleaned['end_date'] = campaign_cleaned['end_date'].dt.date

# Display the DataFrame to verify the changes.
print(campaign_cleaned.head().to_string())
```

```python
# Export the DataFrame as a CSV file. 
campaign_cleaned.to_csv("Resources/campaign.csv", index=False)
```

```python
# Read the data into a Pandas DataFrame. Use the `header=2` parameter when reading in the data.
contact_info_df = pd.read_excel('Resources/contacts.xlsx', header=2)
contact_info_df.head()
```

### Create the Contacts DataFrame 
---
**Create a Contacts DataFrame that has the following columns:**
- A column named "contact_id"  that contains the unique number of the contact person.
- A column named "first_name" that contains the first name of the contact person.
- A column named "last_name" that contains the first name of the contact person.
- A column named "email" that contains the email address of the contact person

Then export the DataFrame as a `contacts.csv` CSV file.

### Option 1: Use Pandas to create the contacts DataFrame.

```python
# Iterate through the contact_info_df and convert each row to a dictionary.
import json
dict_values = []

for i in range(len(contact_info_df)):
    # Skip the first row which is the header
    if i == 0:
        continue
    # Convert the string to a dictionary using json.loads
    contact_dict = json.loads(contact_info_df.iloc[i, 0])
    # Append the values to dict_values
    dict_values.append([contact_dict['contact_id'], contact_dict['name'], contact_dict['email']])

# Print out the list of values for each row.
print(dict_values)
```

```python
# Create a DataFrame from the list of values
contact_info = pd.DataFrame(dict_values, columns=['contact_id', 'name', 'email'])

# Print the DataFrame
print(contact_info.to_string())
```

```python
# Check the data types and other information of each column
contact_info.info()
```

```python
# Split the 'name' column into 'first_name' and 'last_name'
contact_info['first_name'] = contact_info['name'].apply(lambda x: x.split(' ')[0])
contact_info['last_name'] = contact_info['name'].apply(lambda x: ' '.join(x.split(' ')[1:]))

# Drop the 'name' column
contact_info = contact_info.drop('name', axis=1)

# Print the DataFrame
print(contact_info.to_string())
```

```python
# Reorder the columns
contact_info = contact_info[['contact_id', 'first_name', 'last_name', 'email']]

# Print the DataFrame
print(contact_info.to_string())
```

```python
# Check the datatypes one more time before exporting as CSV file.
contact_info.info()
```

```python
# Export the DataFrame as a CSV file.
contact_info.to_csv("Resources/contacts.csv", encoding='utf8', index=False)
```
